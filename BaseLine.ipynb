### Разбиение датасета на тренировочную и тестовую выборки:
Для того, чтобы обеспечить объективные оценки качества модели, мы разбили набор данных на тренировочную и тестовую выборки. Для этого использовали функцию train_test_split из scikit-learn, разделили данные в соотношении 80/20, где 80% данных используются для тренировки модели, 20% - для оценки качества.

```python
# Сделаем разбиение данных на тренировочную и тестовую выборки
X = dataset.drop(columns=['Delivery_person_Age'])  # Признаки
y = dataset['Delivery_person_Ratings']  # Целевая переменная
```

```python
# Разбиение данных на тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
from sklearn.dummy import DummyRegressor
```

```python
d_reg = DummyRegressor(strategy="mean")
d_reg.fit(X_train, y_train)
```

```python
y_predict_dummy = d_reg.predict(X_test)
```

### Измерение качества константного предсказания:

Затем для того, чтобы сделать оценку качества модели мы начали с константного предсказания. Для этого использовали  стратегию "mean", которая считает предсказывает среднее значение. Мы измерили точность константного предсказания с помощью метрики MAE.

```python
from sklearn.metrics import mean_absolute_error
```

```python
#Измерим качества константного предсказания
mae_dummy =  mean_absolute_error(y_test, y_predict_dummy)
print("Mean Squared Error of Dummy Regressor:", mae_dummy)
```
